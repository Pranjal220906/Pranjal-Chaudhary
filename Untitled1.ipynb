1. Creating a Pandas Series

```python
stock_price = [150.25, 155.50, 152.00, 158.75, 160.10, 159.90,
161.50]
```

```python
day = ['Monday',
'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

daily_prices = pd.Series(data=stock_price, index=day)
daily_prices
```

Question 2:
You are a data intern for an online electronics store. Your manager has given you the inventory
data for a new line of smart gadgets. The data is currently stored in a Python dictionary, where
the keys are the product IDs and the values are the number of units in stock. Your task is to use
Pandas Series to efficiently manage and analyze this inventory data.

```python
gadget_inventory = {'P101': 75, 'P102': 120, 'P103': 50, 'P104':
90, 'P105': 15}
```

```python
print (gadget_inventory)
```

```python
ProductIDs= list(gadget_inventory.values())
print(ProductIDs)
```

```python
gadget_series = pd.Series(gadget_inventory)
print(gadget_series.index)
print(gadget_series.values)
print(gadget_series.sum())
```

```python
print()
```

```python
low_stock_products = gadget_series[gadget_series < 60]
print(low_stock_products)
```
