```python
import pandas as pd
```

```python
rides = pd.read_csv("https://raw.githubusercontent.com/pradhrahul/Data-Science-Visualization/refs/heads/main/ride_sharing_clean.csv")
rides.info()
```

```python
rides.head()
```

```python
rides['user_type'] = rides['user_type'].astype(int)
rides['user_type'].info()
```

```python
rides['user_type'].describe()
```

```python
rides['user_type'].astype("category").dtype
```

```python
rides['user_type']= rides['user_type'].astype("category")
```

```python
rides['user_type'].dtype
```

```python
rides['user_type'].describe()
```

```python
dict_ut = {
    1: "Free Ride",
    2: "Pay per ride",
    3: "Monthly Subscriber"
}
```

```python
rides['new_usertype']= rides['user_type'].map(dict_ut)
```
