```python
df.dest_region = df.dest_region.str.lower()
df.head()
```

```python
df.dest_size = df.dest_size.str.strip(' ');
df.head()
```

```python
print(df.dest_size.unique())
```

```python
wait_values = df['wait_min']
df['wait_type'] = ['short' if x in range(0,60) else 'long' for x in wait_values]
df.head()
df.wait_type.unique()
```

```python
day = df.day
df['day_week'] = ['weekend' if x.lower() == 'sunday' else 'weekday' for x in day]
df.day_week.unique()
```

```python
label_ranges = [0,60,180,np.inf]
label_names = ['short','medium','long']
df['wait_type'] = pd.cut(df['wait_min'], bins=label_ranges, labels=label_names,right = False)
df.head()
```

```python

dry_mapping = {
    'Monday': 'weekday',
    'Tuesday': 'weekday',
    'Wednesday': 'weekday',
    'Thursday': 'weekday',
    'Friday': 'weekday',
    'Saturday': 'weekday',
    'Sunday': 'weekend',
}
df['day_week'] = df.day.replace(dry_mapping)
df.head()
```
