##### Broadcastable

* `(3, 4)` and `(1, 4)`
* `(3, 4)` and `(4, )`
* `(3, 4)` and `(3, 1)`

##### Not Broadcastable

* `(3, 4)` and `(1, 2)`
* `(3, 4)` and `(4, 1)`
* `(3, 4)` and `(3, )`
